```python
# Importing libraries
import pandas as pd
import numpy as np

import sklearn
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import learning_curve,LearningCurveDisplay
from sklearn.metrics import r2_score

from xgboost import XGBRegressor
from feature_engine.datetime import DatetimeFeatures

import joblib

import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns',None)
```

```python
sklearn.set_config(transform_output='pandas')
```

```python
#Importing Datasets
train_df = pd.read_csv('data/train.csv')
val_df = pd.read_csv('data/val.csv')
test_df = pd.read_csv('data/test.csv')
```

```python
train_df
```

```python
def split_data(data):
    X = data.drop('price',axis = 1)
    y = data['price']
    return (X,y)
```

```python
X_train , y_train = split_data(train_df)
X_val , y_val = split_data(val_df)
X_test , y_test = split_data(test_df)
```

# Data Preprocessing

```python
dt_cols = ['date_of_journey','dep_time','arrival_time']
num_cols = ['duration','total_stops']
cat_cols = [col for col in X_train.columns if (col not in dt_cols) and (col not in num_cols)]
```

```python
cat_cols
```

```python
num_pipe = Pipeline(steps = [
    ('imputer',SimpleImputer(strategy = 'mean')),
    ('scaler',StandardScaler())
])
                    
cat_pipe = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy = 'most_frequent')),
    ('encoder',OneHotEncoder(sparse_output=False,handle_unknown='ignore'))
])

doj_pipe = Pipeline(steps = [
    ('imputer',SimpleImputer(strategy = 'most_frequent')),
    ('extractor',DatetimeFeatures(features_to_extract=['month','week','day_of_week','day_of_month'],format = 'mixed')),
    ('scaler',StandardScaler())
])

toj_pipe = Pipeline(steps = [
    ('imputer',SimpleImputer(strategy = 'most_frequent')),
    ('extractor',DatetimeFeatures(features_to_extract=['hour','minute'],format = 'mixed')),
    ('scaler',StandardScaler())
])
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num',num_pipe,num_cols),
    ('cat',cat_pipe,cat_cols),
    ('date',doj_pipe,['date_of_journey']),
    ('time',toj_pipe,['dep_time','arrival_time'])
])
```

```python
preprocessor.fit_transform(X_train)
```

```python
# Model Selection 
algorithms = {'Linear Regression':LinearRegression(),
             'Support Vector Machine':SVR(),
              'Random Forest': RandomForestRegressor(n_estimators=10),
              'XG Boost': XGBRegressor(n_estimators = 10)
              }
```

```python
X_data_df = pd.concat([train_df,val_df],axis = 0)
X_data,y_data = split_data(X_data_df)
```

```python
def plot_curves(sizes,mean_scores,std_scores,label,ax):
    ax.plot(sizes,
           mean_scores,
           marker = 'o',
           label=label
           )
    ax.fill_between(x=sizes,
                   y1=mean_scores-std_scores,
                   y2=mean_scores+std_scores,
                   alpha=0.5
                   )
```

```python
def plot_learning_curves(name,algorithm,figsize=[12,5]):
    model = Pipeline(steps = [
        ('pre',preprocessor),
        ('alg',algorithm)
    ])
    
    train_sizes,train_scores,test_scores = learning_curve(estimator=model,X = X_data,y = y_data,
                                                       n_jobs=-1,scoring = 'r2',cv = 3,random_state = 42,error_score='raise')
    
    mean_train_scores = np.mean(train_scores,axis = 1)
    std_train_scores = np.std(train_scores,axis = 1)
    train_score = f"{mean_train_scores[-1]:.2f} +/- {std_train_scores[-1]:.2f}"
    
    mean_test_scores = np.mean(test_scores,axis = 1)
    std_test_scores = np.std(test_scores,axis = 1)
    test_score = f"{mean_test_scores[-1]:.2f} +/- {std_test_scores[-1]:.2f}"
    
    fig,ax = plt.subplots(figsize = figsize)
    
    LearningCurveDisplay(train_sizes=train_sizes,
                              train_scores=train_scores,
                         test_scores=test_scores,score_name= 'r2').plot(ax=ax)
    ax.set(xlabel = 'Train Set Size',ylabel = 'R-Square')
    
    ax.set_title(f"Learning_curves: {name}\n Train Score:{train_score} | Test Score :{test_score}")
    
    handles, labels = ax.get_legend_handles_labels()
    
    ax.legend(handles, labels, loc='best')
    
    plt.show()
```

```python
for name,alg in algorithms.items():
    plot_learning_curves(name,alg)
```

# Model Training

```python
model = Pipeline(steps = [
    ('pre',preprocessor),
    ('rf',RandomForestRegressor(n_estimators=100,max_depth=5))
])
```

```python
model.fit(X_data,y_data)
```

```python
def evaluate_model(X,y,model):
    predicted = model.predict(X)
    return r2_score(y,predicted)
```

```python
print(f'Training Prediction is {evaluate_model(X_data,y_data,model)} and Test Prediction is {evaluate_model(X_test,y_test,model)}')
```
